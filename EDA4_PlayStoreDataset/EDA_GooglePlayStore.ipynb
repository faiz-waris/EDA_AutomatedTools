#### Replacing M with '000' and K with ' ' since it is a numerical data type

```python
df_copy["Size"] = df_copy["Size"].str.replace('M','000')
```

```python
df_copy["Size"]
```

```python
df_copy["Size"] = df_copy["Size"].str.replace('k','')
```

```python
df_copy["Size"].unique()
```

```python
df_copy["Size"]=df_copy["Size"].str.replace("Varies with device", str(np.nan))
```

```python
# Now we can change the data type as float
```

```python
df_copy["Size"].dtype
```

```python
df_copy['Size']=df_copy["Size"].astype("float")
```

```python
df_copy["Size"].head()
```

```python
df_copy["Size"][2]*1000
```

```python
for i in df_copy['Size']:
    if i < 10:
        df_copy['Size']=df_copy['Size'].replace(i,i*1000)
```

```python
df_copy["Size"].head()
```

```python
df_copy["Size"].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy["Price"].unique()
```

```python
# need to replace items with empty string (data cleaning)
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs',"Price"]
for item in chars_to_remove:
    for col in cols_to_clean:
        df_copy[col]=df_copy[col].str.replace(item,'')
```
